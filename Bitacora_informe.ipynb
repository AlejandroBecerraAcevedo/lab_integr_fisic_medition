- **Tuerca**: Se midieron el diámetro externo $(D_{ext})$, diámetro interno $(D_{int})$, altura $(H)$ y la longitud $(L)$ con el pie de rey.

## **Cálculo del Volumen e Incertidumbre**
Para cada objeto, el volumen se calculó según la forma geométrica:

1. **Cubo de madera**:  
   $$
   V = L \times W \times H
   $$
   La incertidumbre del volumen se obtuvo con la propagación de errores:
   $$
   \delta V = V \left( \frac{\delta L}{L} + \frac{\delta W}{W} + \frac{\delta H}{H} \right)
   $$

2. **Cilindro**:  
   Se calculó el radio como $ r = \frac{D}{2} $, y el volumen:
   $$
   V = \pi r^2 h
   $$
   La incertidumbre del volumen fue:
   $$
   \delta V = V \left( 2 \frac{\delta r}{r} + \frac{\delta h}{h} \right)
   $$

3. **Esfera**:  
   Se calculó el radio como $ r = \frac{D}{2} $, y el volumen:
   $$
   V = \frac{4}{3} \pi r^3
   $$
   La incertidumbre del volumen fue:
   $$
   \delta V = V \left( 3 \frac{\delta r}{r} \right)
   $$

4. **Tuerca**:  
   Se calculó el volumen como la diferencia entre dos cilindros:
   $$
   V = \pi H (r_{ext}^2 - r_{int}^2)
   $$
   La incertidumbre del volumen fue:
   $$
   \delta V = V \left( 2 \frac{\delta r_{ext}}{r_{ext}} + 2 \frac{\delta r_{int}}{r_{int}} + \frac{\delta H}{H} \right)
   $$


> **Instrumentos utilizados:**  
> - **Pie de Rey (Calibrador Vernier)**: Para todas las mediciones, excepto la esfera.  
> - **Micrómetro**: Para la esfera.  
> - **Balanza**: Para la masa, con una incertidumbre de ± 0.01 g.  


---

## **Cálculo de la Densidad e Incertidumbre**
La densidad se calculó con la ecuación:

$$
\rho = \frac{m}{V}
$$

La incertidumbre de la densidad se obtuvo propagando los errores:

$$
\frac{\delta \rho}{\rho} = \frac{\delta m}{m} + \frac{\delta V}{V}
$$

Despejando:

$$
\delta \rho = \rho \left( \frac{\delta m}{m} + \frac{\delta V}{V} \right)
$$

# **Cálculo de Volumen, Densidad e Incertidumbres**

## **1. Cubo de Madera**

### Medidas:
- Largo: $L = 58.40 \pm 0.05$ mm  
- Ancho: $A = 39.18 \pm 0.05$ mm  
- Alto: $H = 35.30 \pm 0.05$ mm  
- Masa: $m = 52.24 \pm 0.01$ g  

### Volumen:
$$V = L \cdot A \cdot H = 58.40 \cdot 39.18 \cdot 35.30 = 80760.046\ \text{mm}^3 = 80.760046\ \text {cm}^3$$

### Incertidumbre del volumen:
$$\frac{\delta V}{V} = \frac{\delta L}{L} + \frac{\delta A}{A} + \frac{\delta H}{H} = \frac{0.05}{58.40} + \frac{0.05}{39.18} + \frac{0.05}{35.30} = 0.003549$$\
$$\delta V = V \cdot 0.003549 = 80760.046 \cdot 0.003549 = 286.611\ \text{mm}^3 = 0.286611\ \text {cm}^3$$

### Densidad:
$$\rho = \frac{m}{V} = \frac{52.24}{80.760046} = 0.646855\ \text{g/cm}^3$$

### Incertidumbre de la densidad:
$$\frac{\delta \rho}{\rho} = \frac{\delta m}{m} + \frac{\delta V}{V} = \frac{0.01}{52.24} + \frac{0.286611}{80.760046} = 0.00374$$  
$$\delta \rho = 0.646855 \cdot 0.00374 = 0.00249\ \text{g/cm}^3$$

## **2. Cilindro**

### Medidas:
- Diámetro: $D = 25.15 \pm 0.05$ mm → $r = 12.575 \pm 0.025$ mm  
- Altura: $h = 18.25 \pm 0.05$ mm  
- Masa: $m = 68.60 \pm 0.01$ g  

### Volumen:
$$V = \pi r^2 h = \pi \cdot (1.2575)^2 \cdot 1.825 = 9.0663\ \text{cm}^3$$

### Incertidumbre del volumen:
$$\frac{\delta V}{V} = 2\frac{\delta r}{r} + \frac{\delta h}{h} = 2 \cdot \frac{0.0025}{1.2575} + \frac{0.005}{1.825} = 0.00672$$\
$$\delta V = 9.0663 \cdot 0.00672 = 0.061\ \text{cm}^3$$

### Densidad:
$$\rho = \frac{68.60}{9.0663} = 7.56\ \text{g/cm}^3$$

### Incertidumbre de la densidad:
$$\frac{\delta \rho}{\rho} = \frac{0.01}{68.60} + \frac{0.061}{9.0663} = 0.00687$$\
$$\delta \rho = 7.56 \cdot 0.00687 = 0.052\ \text{g/cm}^3$$

\ \text{cm}^3$$

## **4. Tuerca**

### Medidas:

- **Lado del hexágono:** $b = 9.00 \pm 0.05$ mm $= 0.900 \pm 0.005$ cm  
- **Altura (h):** $8.25 \pm 0.05$ mm $= 0.825 \pm 0.005$ cm  
- **Radio del hueco interior:** $r = 3.70 \pm 0.025$ mm $= 0.370 \pm 0.0025$ cm  
- **Masa:** $m = 8.64 \pm 0.01$ g  

## Cálculo del Volumen de la Tuerca

La tuerca se modela como un **prisma hexagonal regular** con un **cilindro hueco** en su interior.

### 1. Volumen del prisma hexagonal (sin hueco)

Área base del hexágono regular en función del lado:

$$
A_{hex} = \frac{3\sqrt{3}}{2} \cdot b^2 = \frac{3\sqrt{3}}{2} \cdot (0.900)^2 \approx 2.102\ \text{cm}^2
$$

Volumen del prisma:

$$
V_{hex} = A_{hex} \cdot h = 2.102 \cdot 0.825 = 1.735\ \text{cm}^3
$$



### 2. Volumen del cilindro interior (hueco)

$$
V_{cil} = \pi r^2 h = \pi \cdot (0.370)^2 \cdot 0.825 = 0.353\ \text{cm}^3
$$



### 3. Volumen neto de la tuerca

$$
V = V_{hex} - V_{cil} = 1.735 - 0.353 = 1.382\ \text{cm}^3
$$


### Incertidumbre del Volumen

### Incertidumbre relativa del volumen del hexágono

Usamos propagación de incertidumbre:

$$
\frac{\delta V_{hex}}{V_{hex}} = 2 \cdot \frac{\delta b}{b} + \frac{\delta h}{h}$$\

$$
= 2 \cdot \frac{0.005}{0.900} + \frac{0.005}{0.825} = 0.01111 + 0.00606 = 0.01717
$$\

$$
\delta V_{hex} = 1.735 \cdot 0.01717 = 0.030\ \text{cm}^3
$$

### Incertidumbre relativa del volumen del cilindro

$$
\frac{\delta V_{cil}}{V_{cil}} = 2 \cdot \frac{\delta r}{r} + \frac{\delta h}{h} = 2 \cdot \frac{0.0025}{0.370} + \frac{0.005}{0.825} = 0.01351 + 0.00606 = 0.01957
$$\

$$
\delta V_{cil} = 0.353 \cdot 0.01957 = 0.007\ \text{cm}^3
$$

### Incertidumbre total del volumen (propagación)

$$
\delta V = \sqrt{(\delta V_{hex})^2 + (\delta V_{cil})^2} = \sqrt{(0.030)^2 + (0.007)^2} = 0.031\ \text{cm}^3
$$


## Densidad

$$
\rho = \frac{m}{V} = \frac{8.64}{1.382} = 6.25\ \text{g/cm}^3
$$


### Incertidumbre de la Densidad

$$
\frac{\delta \rho}{\rho} = \frac{\delta m}{m} + \frac{\delta V}{V} = \frac{0.01}{8.64} + \frac{0.031}{1.382} = 0.00116 + 0.02243 = 0.02359
$$

$$
\delta \rho = 6.25 \cdot 0.02359 = 0.15\ \text{g/cm}^3
$$

```python
# count dataframe rows number how many  times the number 4 appears in the dataframe and print each row
counts = l == 4
counts.sum(axis=1)
```

## Nuestra distribución de probabilidad binomial esta dada por:
$$ P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k} $$

siendo

$p = \frac{1}{6}$ (Probabilidad de que salga 4 al tirar un dado)  
$n = 9$ (Lanzamientos)  


```python
from math import sqrt

# teórico
n_repeticiones, n_lanzamientos = l.shape
probabilidad_de_4 = 1 / 6

media_teorica = n_lanzamientos * probabilidad_de_4
desviacion_teórica = sqrt(n_lanzamientos * probabilidad_de_4 * (1 - probabilidad_de_4))
print("Esta es la media teórica: ", media_teorica)
print("Esta es la desviación teórica: ", desviacion_teórica)
```

```python
# medido
prob = (counts.sum(1) == 3).sum() / l.shape[0]
desviacion = ((l.shape[0] * prob) * (1 - prob)) ** 0.5
print("Esta es la media probabilística del experimento: ", 20 * 0.05)
print("Esta es la desviación estándar del experimento: ", desviacion)
```

```python
probs = counts.sum(1).value_counts() / n_repeticiones
probs = probs.reindex(np.arange(0, n_lanzamientos + 1), fill_value=0)
probs
```

```python
k = np.arange(0, n_lanzamientos + 1)
plt.bar(k, probs, label="medido")
plt.plot(k, stats.binom.pmf(k, n_lanzamientos, probabilidad_de_4), color="orange", label="teórico")
plt.xlabel("Veces en que sale 4")
plt.ylabel("Probabilidad")
plt.legend()
```
